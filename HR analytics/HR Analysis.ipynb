```python
%%sql

--EMPLOYEES TABLE
SELECT * FROM employees;
```

```python
%%sql

--What is the average age of employees in each department?
SELECT Department, AVG(Age) AS AvgAge
FROM Employees
GROUP BY Department;
```

```python
%%sql

--What is the highest salary among male and female employees?
SELECT Gender, MAX(Salary) AS MaxSalary
FROM Employees
GROUP BY Gender;
```

```python
%%sql

-- How many employees are there in each department?
SELECT Department, COUNT(*) AS NumEmployees
FROM Employees
GROUP BY Department;
```

```python
%%sql

-- Which department has the highest average salary?
SELECT Department, AVG(Salary) AS AvgSalary
FROM Employees
GROUP BY Department
ORDER BY AvgSalary DESC
LIMIT 1;
```

```python
%%sql

--How many employees have more than 10 years of experience?
SELECT COUNT(*) AS ExperiencedEmployees
FROM Employees
WHERE YearsExperience > 10;
```

```python
%%sql

--What is the average rating given in performance reviews?
SELECT AVG(Rating) AS AvgRating
FROM PerformanceReviews;
```

```python
%%sql

--How many performance reviews have been conducted for each employee?
SELECT EmployeeID, COUNT(*) AS NumReviews
FROM PerformanceReviews
GROUP BY EmployeeID;
```

```python
%%sql

--What is the distribution of ratings across performance reviews?
SELECT Rating, COUNT(*) AS NumReviews
FROM PerformanceReviews
GROUP BY Rating
ORDER BY Rating;
```

```python
%%sql

-- Which employee received the highest rating in their performance review?
SELECT EmployeeID, MAX(Rating) AS MaxRating
FROM PerformanceReviews
GROUP BY EmployeeID;
```

```python
%%sql

--How many performance reviews were conducted in each month?

SELECT EXTRACT(MONTH FROM ReviewDate) AS Month, COUNT(*) AS NumReviews
FROM PerformanceReviews
GROUP BY EXTRACT(MONTH FROM ReviewDate)
ORDER BY Month;
```
